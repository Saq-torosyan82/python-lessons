```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

#### 1.1. Скачать следующие данные: kc-house-data и laptop_price

```python
houses = pd.read_csv('kc_house_data.csv', encoding='windows-1251', index_col='id', parse_dates=['date'])
houses.head()
```

#### 1.2. Изучите стоимости недвижимости

```python
houses['price'].min(), houses['price'].max() 
```

```python
plt.figure(figsize=(12, 6))
plt.hist(houses['price'])
plt.title('Стоимости недвижимости')
plt.xlabel('Стоимость')
plt.ylabel('Количество')
plt.xticks(range(0, 8000000, 1000000), rotation=5);
```

#### 1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(8, 6))
plt.hist(houses['sqft_living'])
plt.title('Квадратуры жилой')
plt.xlabel('Квадрат')
plt.ylabel('Количество');
```

#### 1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(8, 6))
plt.hist(houses['yr_built'])
plt.title('Года постройки')
plt.xlabel('Год')
plt.ylabel('Количество');
```

### 2 задача

#### 2.1. Изучите распределение домов от наличия вида на набережную
- Постройте график
- Сделайте выводы

```python
waterfront_data = houses['waterfront'].value_counts()
waterfront_data
```

```python
plt.figure(figsize=(8, 6))
plt.pie(waterfront_data, autopct='%1.2f%%')
plt.title('домов от наличия вида на набережную')
plt.legend(waterfront_data.index);
```

Вывод - много домов у которых нету вид на набережную

#### 2.2. Изучите распределение этажей домов

```python
floors_data = houses['floors'].value_counts()
floors_data
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=floors_data.index, y=floors_data.values)
plt.title('распределение этажей домов')
plt.xlabel('этажи')
plt.ylabel('Количество');
```

#### 2.3. Изучите распределение состояния домов

```python
condition_data = houses['condition'].value_counts()
condition_data
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=condition_data.index, y=condition_data.values)
plt.title('распределение состояния домов')
plt.xlabel('Состояния')
plt.ylabel('Количество');
```
