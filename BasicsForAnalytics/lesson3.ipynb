### Задача 1

- Скачать данные
- Считать данные с помощью pandas
- Вывести на экран первые 5 строк

```python
import pandas as pd
```

```python
houses = pd.read_csv('kc_house_data.csv', encoding='windows-1251', index_col='id', parse_dates=['date'])
houses.head()
```

#### 1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

```python
houses['delta_renovated'] = houses['yr_renovated'] - houses['yr_built']
houses.loc[houses['delta_renovated'] < 0, 'delta_renovated'] = 0
houses.sample(5)
```

#### 1.3 Создайте признаки года продажи, месяца продажи

```python
houses['sale_year'] = houses['date'].dt.year
houses['sale_month'] = houses['date'].dt.month
houses.sample(5)
```

#### 1.4 Удалите признаки date, zipcode, lat, long

```python
houses.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
houses.head()
```

### 3 задача
Составьте несколько сводных таблиц

#### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
Отсортируйте от меньшей стоимости к большей

```python
grouped = houses.groupby('bedrooms')
grouped.agg({'price': 'mean'}).sort_values('price')
```

#### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома

```python
grouped = houses.groupby('condition')
grouped.agg({'price': ['min', 'mean', 'max']})
```

#### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
houses.pivot_table(index='waterfront', columns='view', aggfunc='count', values='price', fill_value=0)
```

#### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
houses.pivot_table(index='floors', columns='bedrooms', aggfunc='count', values='price', fill_value=0)
```

#### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
pd.crosstab(index=houses['condition'], columns=houses['grade'], values='price', aggfunc='mean')
```
